### Â¿En quÃ© plataformas empezaste o acabaste los cursos de ciencia de datos?

```python
primer_metodo = (
    encuesta
    .filter(regex='Q6_|id_usuario')
    .melt(
        id_vars='id_usuario',
        value_vars=None,
        var_name='pregunta',
        value_name='respuesta')
    .dropna()
)
primer_metodo.sample(10)
```

```python
(
    primer_metodo
    .value_counts('respuesta',normalize=True)
    .rename("nro_usuarios")
    .reset_index()
)
```

### Â¿Otro mÃ©todo para realizar el mismo anÃ¡lisis?

```python
analisis_agregado = (
    primer_metodo
    .groupby(['respuesta'])
    .agg(nro_usuarios=('id_usuario','count'))
    .reset_index()
    .sort_values('nro_usuarios',ascending=False)
    .assign(
        porc = lambda x: x.nro_usuarios/x.nro_usuarios.sum()
      )
)
analisis_agregado
```

### Â¿Puedo darle formato a los decimales?

```python
(
    analisis_agregado
    .assign(
        porc = lambda col: col.porc
                  .apply(
                      lambda x: "{0:.2f}%".format(x*100)
                      )
    )
)
```
